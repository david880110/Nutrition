```python
# Ddd dependencies
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.orm import Session
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

from sqlalchemy import Column, Integer, String, Float, Date
```

```python
# Load csv files
rm_data = 'raw_data/raw_material.csv'
fp_data = 'raw_data/finished_product.csv'
```

```python
# Read Raw Data File and store into Pandas Data Frames
rm_df = pd.read_csv(rm_data)
fp_df = pd.read_csv(fp_data)
```

```python
# Create the Schools_data class
class Raw_material_table(Base):
    __tablename__ = 'raw_material_table'
    article_no = Column(Integer, primary_key = True)
    dataset = Column(String(255))
    category = Column(String(255))
    vendor = Column(String(255))
    article_description_en = Column(String(255))
    article_description_ch = Column(String(255))
    pack_size = Column(String(255))
    source = Column(String(255))
    serving_size = Column(Float)
    total_calories_cal = Column(Float)
    calories_from_fat_fat_cal = Column(Float)
    total_fat_g = Column(Float)
    saturated_fat_g = Column(Float)
    trans_fat_g = Column(Float)
    cholesterol_mg = Column(Float)
    sodium_mg = Column(Float)
    total_carbohydrate_g = Column(Float)
    dietary_fiber_g = Column(Float)
    sugars_g = Column(Float)
    protein_g = Column(Float)

class Finished_product_table(Base):
    __tablename__ = 'finished_product_table'
    index = Column(Integer, primary_key = True)
    section = Column(String(255))
    toledo_number = Column(Float)
    type = Column(String(255))
    english_description = Column(String(255))
    chinese_description = Column(String(255))
    serving_size = Column(String(255))
    unit = Column(String(255))
    total_calories_cal = Column(Float)
    percent_total = Column(Float)
    calories_from_fat_fat_cal = Column(Float)
    total_fat_g = Column(Float)
    percent_fat = Column(Float)
    saturated_fat_g = Column(Float)
    percent_satfat = Column(Float)
    trans_fat_g = Column(Float)
    cholesterol_mg = Column(Float)
    percent_chol = Column(Float)
    sodium_mg = Column(Float)
    percent_sodi = Column(Float)
    total_carbohydrate_g = Column(Float)
    percent_carb = Column(Float)
    dietary_fiber_g = Column(Float)
    percent_diet = Column(Float)
    sugars_g = Column(Float)
    protein_g = Column(Float)
    percent_prot = Column(Float)
```

```python
# Create the datbase and add the classes
engine = create_engine("sqlite:///nutrition_database.sqlite", echo = False)
conn = engine.connect()
session = Session(engine)

Base.metadata.create_all(conn)
```

```python
# Send the station dataframe to the sqlite database
rm_df.to_sql('raw_material_table', engine, if_exists='replace')
fp_df.to_sql('finished_product_table', engine, if_exists='replace')
```
